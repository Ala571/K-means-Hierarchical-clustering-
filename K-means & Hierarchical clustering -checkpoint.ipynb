# Data preprocessing

```python
import pandas as pd 
data=pd.read_csv("CC GENERAL.csv")
data
```

```python
data.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
def plot_correlation_map( df ):
    

    corr = df.corr()

    s , ax = plt.subplots( figsize =( 20 , 20) )

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12 }

        )

plot_correlation_map(data)
```

```python
data.update((data['MINIMUM_PAYMENTS']).fillna(value=(data['MINIMUM_PAYMENTS']).mean(), inplace = True))
```

```python
data.update((data['CREDIT_LIMIT']).fillna(value=(data['CREDIT_LIMIT']).mean(), inplace = True))
```

```python
data.isnull().sum().sum()
```

```python
data.drop('CUST_ID',
  axis='columns', inplace=True)
```

# Agglomerative_Clustering

```python
from sklearn.cluster import AgglomerativeClustering #Importing our clustering algorithm : Agglomerative
model=AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='complete')
clust_labels=model.fit_predict(data)  #Applying agglomerative algorithm with 5 clusters, using euclidean distance as a metric
```

```python
agglomerative=pd.DataFrame(clust_labels)  #rendre clust_labels dataframe 
agglomerative
```

```python
# Each colored group represents a cluster.
import matplotlib.pyplot as plt
fig =plt.figure()
ax = fig.add_subplot(111)
scatter = ax.scatter (data ['CREDIT_LIMIT'] , data ["MINIMUM_PAYMENTS"] , c= agglomerative[0], s=50)
#afficher d'apres le credit limit et le minimum payements agglomerative clustering 
ax.set_title("Agglomerative Clutering")

plt.colorbar(scatter)

```

```python
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(20,20))
plt.title("Dendrograms")
dend=shc.dendrogram(shc.linkage(data, method="complete"))

#afficher le dendogram du data 
```

```python
from sklearn.cluster import KMeans  #Importing our clustering algorithm: KMeans
kmeans=KMeans(n_clusters=5, random_state=0)  #Cluster our data by choosing 5 as number of clusters
kmeans.fit(data) #fitting the data 
```

```python
labels=pd.DataFrame(kmeans.labels_)
labels
```

```python
kmeans.predict(data)
print(kmeans.cluster_centers_)
```

```python
distance=[]
k = range (1,15)
for i in k :
    km= KMeans(n_clusters=i)
    km=km.fit(data)
    distance.append(km.inertia_)
    

plt.plot(k,distance)
```
